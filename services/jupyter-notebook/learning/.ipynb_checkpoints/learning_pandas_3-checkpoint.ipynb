## Data Frame

Pandas series introduced previously has a 1-dimensional data structure. By stacking series together, a 2-dimensional data structure, namely a data frame, can be obtained. Each column of a data frame can be taken as a series.


An example of a data frame is given below.

```python
pokemon_df
```

The above pokemon data frame is imported from a csv table without specifying index. Pandas has created a numeric incremenal index 0, 1, ..., 799 along each row in the table.

Notice that `NaN` is presented for column `Type 2` of some of the rows, which implies that the associated pokemon has only one type. In reality, a `NaN` in a dataset often means missing data. 

When that happens, the associated row usually needs to be handled specifically in one of the following ways:
- Derive the missing data from other known data, if possible.
- Remove the entire row.
- Interpolate the missing data.
- Assign a special flag that says "this row contains a missing data", and handle carefully in later data processing stages.
- Do nothing, if it is safe to do so.

The same applies to pokemon data frame. If we want to analyze the secondary pokemon type of all pokemons, those without a secondary pokemon type rows need to be handled smartly.

It is worth mentioning that **pandas assigns different types do each columns depending on the data it reads, and this might be affected if `NaN` presents in that column**. For example, an integer column with any `NaN` will be taken as a float column. The same applies to pandas series.

### Commonly Used Data Frame Attributes and Methods

Pandas series and data frame share some common attributes and methods, and both of them have unique attributes and methods as well. Commonly used data frame attributes are summarized in the table below.

| Syntax | Description | Also in series |
|:------ |:----------- |:--------------- |
| .size | The length of the data frame. | Y |
| .shape | The shape (length in each dimension) of the data frame. | Y |
| .is_unique | A flag indicating whether there are duplications of values in the series. |
| .values | The values of the data frame, in a 2D array. | Y |
| .index | The index of the data frame. | Y |
| .columns | The column headers given as an index object. | N |
| .axes | A list of index that marks each dimension of the data set. In the case of data frames it is `.index` combined with `.columns`. | D |
| .dtypes | The data type of each column of the data frame. | D |

Commonly used data frame methods are summarized in the table below. More details of these methods are discussed sequentially in this notebook sheet.

| Syntax | Description | Also in series |
|:------ |:----------- |:--------------- |
|.head(n=5)]), .tail(n=5)| Display the first and last rows in the data frame. | Y |
| .info() | A summary of the data frame. | Y |
| .count() | The number of items. |
| .sum() | Sum of all items. |
| .product() | Product of all items. |
| .mean() | Mean of all items. |
| .std() | Standard deviation of all items. |
| .median () | Median of all items. |
| .mode() | The mode of all items. |
| .min(), .max() | Min and max of all items. |
| .value_counts() | Group the values and count the number in each group. Useful when the values are of categories. |
| .sort_values(ascending = True, inplace = False) | Sort the values (the indices moves along with the values). |
| .sort_index(ascending = True, inplace = False) | Sort the index (the values move along with the indices). |
| .copy() | Ducplicate a series decoupled from its origin. |
| .get() | Get a particular item from the series using the index. |
| .apply() | Apply a function to each item in a series. |
| .map() | Map the value of each item in a series using an association. |

Notice that in the above tables "Also in series" column:
- Y: series and data frame have attribute/method of the same name with similar returns.
- D: series and data frame have attribute/method of the same name with different returns.
- N: it is unique to data frame.

To check whether an item is in the index or values of a series, just use `in` as in Python list.
