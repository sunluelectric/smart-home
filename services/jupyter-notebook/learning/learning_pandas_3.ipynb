### Build/Update a Data Frame

A data frame is usually built using one of the following ways:
- Import a data frame from external.
- Build a data frame from scratch or form series.
- Build a data frame on top of an existing data frame by adding new columns and rows.
- Build a data frame from existing data frame(s) by modifying or combining them together.


Add a new constant column to an existing data frame as follows.
```python
<df>.insert(loc = <location>, column = "<new-column-name>", value = <value>)
```
and the new column will be the `<location>`-th column in the data frame, with the original `location`-th column is pushed to the right.

Alternatively, simply use
```python
<df>["<new-column-name>"] = <value>
```
to append a new column to the right of the existing data frame. The assigned `<value>` can be a constant scalar value or a series with approperate length.

Examples are given below.

````python
pokemon_df = pd.read_csv('pokemon.csv')
pokemon_df.insert(loc = 2, column = "In PokeDex", value = "Yes")
pokemon_df
````

````python
pokemon_df["Favourite"] = "Yes"
pokemon_df
````

Add a new dynamic column to an existing data frame as follows. Here "dynamic" means that the content to be added in each row of this new column is not a constant, but depends on its the values in other columns. THis can be achieved by combining adding constant column with pandas series functions such as `.apply()` and `.map()`.

An example is given as follows. Consider adding a "Tier" column to the pokemon data frame. A tier of a pokemon is associated with its "Total" value.

````python
pokemon_df = pd.read_csv('pokemon.csv')

def check_tier(total_sp):
    if total_sp >= 700:
        return "S+"
    elif total_sp >= 600:
        return "S"
    elif total_sp >= 500:
        return "A"
    elif total_sp >= 400:
        return "B"
    elif total_sp >= 300:
        return "C"
    else:
        return "D"

pokemon_df.insert(
    loc = 4,
    column = "Tier",
    value = pokemon_df["Total"].apply(check_tier)
)
pokemon_df
    
````

### Handling Missing Data

Notice that `NaN` is presented for column `Type 2` of some of the rows, which implies that the associated pokemon has only one type. In reality, a `NaN` in a dataset often means missing data. 

When that happens, the associated row usually needs to be handled specifically in one of the following ways:
- Derive the missing data from other known data, if possible.
- Remove the entire row.
- Interpolate the missing data.
- Assign a special flag that says "this row contains a missing data", and handle carefully in later data processing stages.
- Do nothing, if it is safe to do so.

The same applies to pokemon data frame. If we want to analyze the secondary pokemon type of all pokemons, those without a secondary pokemon type rows need to be handled smartly.

Drop rows with any NaN value as follows.
```
<df> = <df>.dropna()
```
where notice that `<df>.dropna()` by iteslf does not modify the data frame.

Drop rows with all its column values NaN as follows.
```
<df> = <df>.dropna(how = "all")
```

Drop rows whose selected columns values have NaN as follows.
```
<df> = <df>.dropna(subset = ["column", ...])
```
in which case if any values with the selected columns is NaN, the row would be removed.

An example is given below.

````python
pokemon_df = pd.read_csv('pokemon.csv')
pokemon_df.dropna()
````

````python
pokemon_df
````

````python
pokemon_df = pokemon_df.dropna()
pokemon_df
````

### Commonly Used Data Frame Attributes and Methods

Pandas series and data frame share some common attributes and methods, and both of them have unique attributes and methods as well. Commonly used data frame attributes are summarized in the table below.

| Syntax | Description |
|:------ |:----------- |
| .size | The total length (column number multiplied by row number) of the data frame. |
| .shape | The shape (length in each dimension) of the data frame. |
| .values | The values of the data frame, in a 2D array. |
| .index | The row names of the data frame. |
| .columns | The column names given as an index object. |
| .axes | The `.index` combined with `.columns`. |
| .dtypes | The data type of each column of the data frame. |

Commonly used data frame methods are summarized in the table below.

| Syntax | Description |
|:------ |:----------- |
|.head(n=5)]), .tail(n=5)| Display the first and last rows in the data frame. |
| .info() | A summary of the data frame. |
| .count() | The number of no-NaN items in each column. |
| .sum(axis = 0) | Sum of all rows (axis 0) or columns (axis 1) and put the result in a series. |
| .value_counts() | Group the values in selected columns tuple and count the number in each group. |
| .insert() | Insert a column to the existing data frame. |
| .sort_values(by = None, ascending = True, inplace = False) | Sort the values by selected columns (the indices moves along with the values). |
| .sort_index(ascending = True, inplace = False) | Sort the index (the values move along with the indices). |
| .copy() | Ducplicate a series decoupled from its origin. |
| .get(\[ "\<column-name\>", ... \]) | Get a sub data frame of selected columns. |

Notice that some of the attributes/methods share the same name and similar purpose between pandas series and data frame, but the returns can have different data structures and types.

### Access Items in Data Frame

To get a sub data frame from a data frame with selected columns, use either the `.get()` function, or use
```python
<df>[["<column-nam1>", ...]]
```

Examples are given below.

````python
pokemon_df[["Name", "Type 1", "Type 2"]]
````

````python
pokemon_df[["Name"]]
````

Notice that `<df>[["column-name"]]` returns a single-column data frame, while `<df>["column-name"]` returns a series. To convert the single-column data frame into a series, use the `.squeeze()` method. An example is given below.

````python
pokemon_df[["Name"]].squeeze()
````
