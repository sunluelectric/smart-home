## Data Importing

Pandas provides flexible data importing tools for variety of data resources types, including csv, excel, clipboard, json, html, xml, latex table, and many more. Details can be found [here](https://pandas.pydata.org/docs/reference/io.html "https://pandas.pydata.org/docs/reference/io.html").

In this notebook series, we mainly focus on importing data from csv files using `.read_csv()` method.


More details about data frame is given in future notebook sheets.

Examples of using `.read_csv()` to import data is given below. 

In the first example, `pokemon.csv` from [here](https://www.kaggle.com/datasets/abcsds/pokemon "https://www.kaggle.com/datasets/abcsds/pokemon") is used as the data resource. It contains a table of many pokemons as well as their types and base parameters, as shown below.

````python
pokemon_df = pd.read_csv('pokemon.csv')
pokemon_df
````

It is possible to import selected columns only when using `.read_csv()`. An example is given below.

````python
pokemon_lite_df = pd.read_csv('pokemon.csv', usecols = ['Name', 'Type 1', 'Type 2'])
pokemon_lite_df
````

It is possible to specify the a column to be used as the labels, instead of using increment labels 0, 1, 2, ... automatically generated by pandas. An example is given below.

````python
pokemon_lite_df = pd.read_csv('pokemon.csv', index_col = 'Name', usecols = ['Name', 'Type 1', 'Type 2'])
pokemon_lite_df
````

A pandas data frame is a stack of pandas series. Each column of the data frame is a pandas series. More details about pandas series is given in future notebook sheets.

A pandas series can be derived from a data frame as follows.

```python
series = dataframe.iloc[:,<column-index>]
series = dataframe['<column-name>']
series = dataframe[['<column-name>']].squeeze('columns')
```
where notice that `dataframe[['<column-name>']]` by itself is a single column data frame, not a series.

Examples are given below.

````python
pokemon_name = pokemon_df.iloc[:,1]
pokemon_name
````

````python
pokemon_name = pokemon_df['Name']
pokemon_name
````

````python
pokemon_name = pokemon_df[['Name']].squeeze('columns')
pokemon_name
````

It is possible to import a series from a csv file using `.read_csv()`, by combining the information above. An example is given below.

````python
pokemon_name_series = pd.read_csv('pokemon.csv', usecols = ['Name']).squeeze('columns')
pokemon_name_series
````
