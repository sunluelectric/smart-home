## Series

Series is a data structure defined in pandas as follows. More details can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.Series.html "https://pandas.pydata.org/docs/reference/api/pandas.Series.html").

```python
class pandas.Series(data=None, index=None, dtype=None, name=None, copy=False, fastpath=False)
```

Pandas series is like a combination of Python list and dictionary: it traces items using both the preserved sequence and also the index.

### Build Series

Pandas series can be created from a Python list or a dictionary as follows.

```python
<series-name> = pd.Series(<values-list>, <index-list>)
<series-name> = pd.Series(<dictionary>)
```

Examples are given below.

````python
s1 = pd.Series(["apple", "banana", "chocolate", "donut", "egg"], [1, 2, 3, 4, 5])
s1
````

````python
s2 = pd.Series(np.random.randn(5), ["a", "b", "c", "d", "e"])
s2
````

From the above example, it is clear that the order of the original lists preserves.

The data type of the series depends on the values. Some commonly seen data types are `float64`, `int64`, `bool` and `object`. Items with different data types can be fixed in a list, in which case the series data type is likely to be identified as `object`. 

If no index is assigned, numeric index will be be used. An example is given below.

````python
s3 = pd.Series(np.random.randn(10))
s3
````

A series can also be created from a Python dictionary. A basic example is given below.

````python
d4 = {
    1: "apple",
    2: "banana",
    3: "chocolate",
    4: "donut",
    5: "egg"
}
s4 = pd.Series(d4)
s4
````

````python
d5 = {
    "a": 2,
    "b": 3,
    "c": 5,
    "d": 7,
    "e": 11
}
s5 = pd.Series(d5)
s5
````

One thing to notice here is that although in the above example, the pandas series looks like a Python dictionary, it indeed assigns hidden numeric orders to the items in the series, hence its order preserves. This makes a big difference between pandas series and Python dictionary.

Due to this reason, it is possible to assign multiple items with the same index in pandas series, which is impossible in Python dictionary.

### Access Items in Series

Use `<series>.get(<index>)` to obtain the item of a series using the index. Alternatively, the items in a series can be accessed from its index as follows.
```python
<series>[index]
```
This allows not only get but also update the items in the series.


When numeric integer index is used in a series, it is possible to retrieve data batch. When non-numeric index is used in a series, it is possible to access the data via its hidden index 0, 1, 2, ... 

Examples are given below.

````python
print("The first item: {}".format(s4[1]))
print("The fifth item: {}".format(s4[5]))
print(s4[[1, 2, 3, 4, 5]]) # return a series
````

````python
print(s4.loc[1:5]) # return batch data in a series
print(s4.loc[:3]) # return batch data in a series from first item
print(s4.loc[3:]) # return batch data in a series till last item
````

````python
print("The first item: {}".format(s5["a"]))
print("The fifth item: {}".format(s5["e"]))
print(s5[["a", "b", "c", "d", "e"]]) # return a series
print("The first item: {}".format(s5[0]))
print("The fifth item: {}".format(s5[4]))
print(s5[[0, 1, 2, 3, 4]]) # return a series
````

````python
print(s4.loc[0:4]) # return batch data in a series
print(s4.loc[:2]) # return batch data in a series from first item
print(s4.loc[2:]) # return batch data in a series till last item
````

With `<series>[index]`, the value in the series can be updated by simply assigning a new value. An example is given below.

````python
print(s4)
s4["5"] = "eggplant"
print(s4)
````

If the index does not exist when assigning, a new item is appended to the series. An example is given below.

````python
s4["6"] = "fries"
print(s4)
````

### Copy of Series

By assigning a series to a new variable, the series is not duplicated. An example is shown below. Although the update is done via the copied series `s4_copy`, the original series `s4` is also affected by the update. 

This implies that similar with C/C++ and Python, the name of the array/list is nothing more than a memory address sharing of the same array/list. By using `=` to assign a new name for a series, the series is not duplicated in the memory.

````python
s4 = pd.Series(d4) # reset s4 from d4
print("Original series: {}".format(s4[1]))
s4_view = s4 # not a true copy, but more than a view
print("Copied series: {}".format(s4_view[1]))
s4_view[1] = "avocado"
print("Copied series after update: {}".format(s4_view[1]))
print("Original series after change: {}".format(s4[1]))
````

To create a true copy, use `.copy()` function as shown in the following example. This creates an independent series decoupled from the original series.

````python
s4 = pd.Series(d4) # reset s4 from d4
print("Original series: {}".format(s4[1]))
s4_copy = s4.copy() # true copy
print("Copied series: {}".format(s4_copy[1]))
s4_copy[1] = "avocado"
print("Copied series after update: {}".format(s4_copy[1]))
print("Original series after change: {}".format(s4[1]))
````

### Apply Method

The `.apply()` method applies a function to each item in a series. An example is given below.

````python
pokemon_series = pd.read_csv("pokemon.csv", usecols = ["Name"]).squeeze("columns")
pokemon_series.apply(len)
````

It is also possible to apply customized function using the `.apply()` method. An example is given below.

````python
def is_long_name(pokemon_name):
    if len(pokemon_name) >= 10:
        return "long"
    elif len(pokemon_name) >= 8:
        return "middle"
    else:
        return "short"
````

````python
pokemon_name_evaluation = pokemon_series.apply(is_long_name)
pokemon_name_evaluation
````

### Map Method

The `.map()` function maps the values in a series using a given association (such as a dictionary, or a pandas series). An example is given below.

````python
mappings = {
    "long" : "l",
    "middle": "m",
    "short": "s"
}

pokemon_name_evaluation.map(mappings)
````

If the mapping cannot be found in the association, a `NaN` will be returned for that item.
