## Series
Pandas defines a data structure called "series" as follows. Details are given [here](https://pandas.pydata.org/docs/reference/api/pandas.Series.html "https://pandas.pydata.org/docs/reference/api/pandas.Series.html").
```python
class pandas.Series(data=None, index=None, dtype=None, name=None, copy=False, fastpath=False)
```

The series class is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index. From this perspective, the series class combines features from both list and dictionary data structures of Python.

### Basics
A series can be created from a Python list. A basic example is given below.

````python
s1 = pd.Series(["apple", "banana", "chocolate", "donut", "egg"], index=[1, 2, 3, 4, 5])
print(s1)
s2 = pd.Series(np.random.randn(5), index=["a", "b", "c", "d", "e"])
print(s2)
````

From the above example, it is clear that the order of the original sequence preserves, and each of them is assigned with an index. 

The data type of the series depends on the data in the list. Some commonly seen data types are `float64`, `int64`, `bool` and `object`. Items with different data types can be fixed in a list, in which case the series data type is likely to be identified as `object`. 

If no index is assigned, numeric index will be be used. An example is given below.

````python
s3 = pd.Series(np.random.randn(10))
print(s3)
````

A series can also be created from a Python dictionary. A basic example is given below.

````python
d4 = {
    1: "apple",
    2: "banana",
    3: "chocolate",
    4: "donut",
    5: "egg"
}
s4 = pd.Series(d4)
print(s4)
d5 = {
    "a": 2,
    "b": 3,
    "c": 5,
    "d": 7,
    "e": 11
}
s5 = pd.Series(d5)
print(s5)
````

One thing to notice here is that although in the above example, the pandas series looks like a Python dictionary, it indeed assigns hidden numeric orders to the items in the series, hence its order preserves. This makes a big difference between pandas series and Python dictionary.

Due to this reason, it is possible to assign multiple items with the same index in pandas series, which is impossible in Python dictionary.

### Commonly Seen Methods and Attributes of Series
Details of pandas series methods and attributes can be found [here](https://pandas.pydata.org/docs/reference/series.html "https://pandas.pydata.org/docs/reference/series.html"), some which are introduced as follows.

Commonly seen methods are listed as follows.

| Syntax | Description |
|:------ |:----------- |
| `.sum()` | Sum of all items. |
| `.product()` | Product of all items. |
| `.mean()` | Mean of all items. |

Commonly seen attributes are aso follows.

| Syntax | Description |
|:------ |:----------- |
| `.size` | The length of the series. |
| `.is_unique` | A flag whether there is any duplication of values (not index) in the series. |
| `.values` | The values of the series, with adaptive data type. |
| `.index` | The index of the series, with adaptive data type. |
| `.dtype` | The data type of the values. |

An example is given below.

````python
s3 = pd.Series(np.random.randn(10))
print("Series: ")
print(s3)
print("Sum of the series is {}".format(s3.sum()))
print("Product of the series is {}".format(s3.product()))
print("Mean of the series is {}".format(s3.mean()))
print("Series size: ")
print(s3.size)
print("List of values and its type: ")
print(s3.values)
print(type(s3.values))
print("Type of each value: ")
print(s3.dtype)
print("List of index and its type: ")
print(s3.index)
print(type(s3.index))
````
